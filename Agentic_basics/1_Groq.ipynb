```python
from groq import Groq

client = Groq()
completion = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
      {
        "role": "system",
        "content": "You are an expert in maths and only and only provide the answer to the question related to maths"
      },
      {
        "role": "user",
        "content": "What is the capital of France?"
      }
    ],
    temperature=1,
    max_completion_tokens=8192,
    top_p=1,
    reasoning_effort="medium",
    stream=True,
    stop=None
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```

```python
from groq import Groq

class ChatGroqWrapper:
    def __init__(self, model="openai/gpt-oss-120b", **defaults):
        self.client = Groq()
        self.model = model
        self.defaults = defaults

    def invoke(self, messages, stream=False):
        return self.client.chat.completions.create(
            model=self.model,
            messages=messages,
            stream=stream,
            **self.defaults
        )

chat = ChatGroqWrapper(
    temperature=1,
    max_completion_tokens=8192,
    top_p=1,
    reasoning_effort="medium"
)

response = chat.invoke(messages, stream=True)

for chunk in response:
    print(chunk.choices[0].delta.content or "", end="")
```
